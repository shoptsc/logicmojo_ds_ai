#### While Loop

```python
counter = 0

while (counter < 10):
    counter = counter + 1
    print(counter)
```

```python
n = int(input('input the positive integer: '))

if n < 0:
    print('please enter a positive integer')
else:
    print(f'Enter number  = {n}')
    num_of_digits = 0
    
    while n > 0:
        n = n//10
        num_of_digits += 1
        
    print(f'There are {num_of_digits} numbers of digits')
```

### Range function in python

`range(start, stop, step)`

```python
x = range(10)
print(x)
```

### For Loop

```python
for i in x:
    print(i)
```

### Nested Loops

```python
for i in range(1, 6):
    print(f'The multiplication table of {i}\n')
    for j in range(1, 11):
        print(f'{i} * {j} = {i * j}')
```
