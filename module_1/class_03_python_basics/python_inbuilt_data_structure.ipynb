## Dictionary

`{key_1 : value_1, 
key_2 : value_2,
key_3 : value_3}`

`{}` -> Empty dictionary

`set()` -> Empty set

```python
fruits = {'apple' : 80, 'banana' : 200, 'mango' : 250, 'orange' : 400}
fruits
```

```python
fruits.items()
```

```python
for k,v in fruits.items():
    print(f'The price of {k} is {v}')
```

```python
fruits.keys()
```

```python
fruits.values()
```

```python
fruit_names = ['mango', 'guava', 'pineapple', 'orange', 'apple']
price_list = [400, 200, 350, 150, 375]

fruit_dict = {}

for x, y in zip(fruit_names, price_list):
    if x not in fruit_dict:
        fruit_dict[x] = y
        
fruit_dict
```

# Dictionary methods

```python
fruit_dict.get('appl', 'it does not exist')
```

```python
'guava' in fruit_dict
```

```python
fruit_dict.pop('apple') # remove the key pair from the dictionary
```
